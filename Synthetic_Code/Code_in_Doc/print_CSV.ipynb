```python
import matplotlib.pyplot as plt
import pandas as pd

# c d liu t tp "gdp_per_capita.csv"
gdp_per_capita = pd.read_csv("gdp_per_capita.csv", thousands=',', delimiter=',',
                             encoding='latin1', na_values="n/a")

# Ly cc ct "Country" v "2015" t DataFrame
countries = gdp_per_capita["Country"]
gdp_2015 = gdp_per_capita["2015"]

# To mt hnh v mi vi kch thc 10x6 inches
plt.figure(figsize=(50, 50))

# V  th
plt.bar(countries, gdp_2015)
plt.xlabel("Country")
plt.ylabel("GDP per capita (2015)")
plt.xticks(rotation=90)
plt.show()
```
